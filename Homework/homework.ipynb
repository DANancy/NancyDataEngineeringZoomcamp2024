```python
# load green zone data
import pandas as pd
from time import time
from sqlalchemy import create_engine

# PostgreSQL connection string
connection_string = 'postgresql://root:root@localhost:5432/ny_taxi'
engine = create_engine(connection_string)

trips_file = 'green_tripdata_2019-09.csv.gz'
table_name = 'green_taxi_trips'

df = pd.read_csv(trips_file, compression='gzip')
df.head(n=0).to_sql(name=table_name, con=engine, if_exists='replace')

df_iter = pd.read_csv(trips_file,compression='gzip', iterator=True, chunksize=100000)

for chunk_df in df_iter:
    t_start = time()
    chunk_df.lpep_pickup_datetime = pd.to_datetime(chunk_df.lpep_pickup_datetime)
    chunk_df.lpep_dropoff_datetime = pd.to_datetime(chunk_df.lpep_dropoff_datetime)
    chunk_df.to_sql(name=table_name, con=engine, if_exists='append')
    t_end = time()
    print('inserted another chunk..., took %.3f second' % (t_end - t_start))
```
